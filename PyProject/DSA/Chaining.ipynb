```python
class Node:
    def __init__(self,key,value):
        self.key= key
        self.value = value
        self.nxtaddress= None
```

```python
class LinkedList:
 
    def __init__(self):
        # Empty Linked list
        self.head = None
        # No. of nodes
        self.nodes = 0

    def __len__(self):
        return self.nodes
    
    def add(self,key,value):
        new_node = Node(key,value)

        if self.head == None:
            self.head = new_node
        else:
            curr = self.head
            while curr.nxtaddress!=None:
                curr = curr.nxtaddress
            curr.nxtaddress = new_node    

    
    def insert_head(self,value):
        # create a new node
        new_node = Node(value)
        # create connection
        new_node.nxtaddress = self.head 
        # assign the node nxtaddress to new head nxtaddress of LL
        self.head = new_node
        # increament nodes no.
        self.nodes = self.nodes + 1

    def traverse(self):
        curr = self.head
        while curr != None:
            print(curr.key,"-->", curr.value,'  ', end=" ")
            curr = curr.nxtaddress

    def size(self):
        curr = self.head
        counter = 0
        while curr != None:
            counter+=1
            curr = curr.nxtaddress
        return counter

    def __str__(self):
        curr = self.head
        result = ' '
        while curr != None:
            result = result + str(curr.data) + '->'
            curr = curr.nxtaddress

        return result[:-2]
    
    def append(self, value):
        new_node = Node(value)
        
        # If list is empty
        if self.head == None:
            self.head = new_node
            self.nodes = self.nodes +1
            return
        
        curr = self.head
        while curr.nxtaddress != None:
            curr = curr.nxtaddress
        # The current nxtaddress id of the last node i.e tail.
        curr.nxtaddress = new_node
        self.nodes = self.nodes + 1

    def insert_after(self,after,value):
        new_node = Node(value)

        curr = self.head
        while curr != None:
            if curr.data == after:
                break
            curr = curr.nxtaddress
        if curr!=None:
            new_node.nxtaddress = curr.nxtaddress
            curr.nxtaddress = new_node
            self.nodes = self.nodes +1
        else:
            return 'Item not found'
        
    def clear(self):
        self.head = None
        self.nodes = 0

    def delete_head(self):
        if self.head == None:
            return 'Empty List!'
        self.head = self.head.nxtaddress
        self.nodes = self.nodes-1 

    def pop(self):
        # when LinkedList have no Item 
        if self.head == None:
            return 'Empty List'
        curr = self.head
        
        # when LinkedList have only one Item (i.e Head)
        if curr.nxtaddress == None:
            return self.delete_head()

        # Reaching upto secondlast list and storing it's address to 'curr'.
        while curr.nxtaddress.nxtaddress != None:
            # transversing the control through the list
            curr = curr.nxtaddress
        # Making the last address to NONE 
        curr.nxtaddress = None   
        self.nodes = self.nodes-1

    def remove(self, key):
        if self.head == None:
            return 'Empty LinkedList'
        
        # When Value to be deleted is Head itself 
        if self.head.key == key:
            return self.delete_head()

        curr = self.head
        while curr.nxtaddress != None:
            if curr.nxtaddress.key == key:
                break
            curr = curr.nxtaddress
        if curr.nxtaddress == None:
            return 'Value Not found in the List'  
        else:
            curr.nxtaddress = curr.nxtaddress.nxtaddress 

    def search(self,key):
        # search method returns the Index no. of the Value
        curr = self.head
        pos = 0
        while curr!= None:
            if curr.key == key:
                return pos
            curr = curr.nxtaddress
            pos = pos + 1
        return -1    
    
    def __getitem__(self,index):
        curr = self.head
        pos = 0
        while curr!= None:
            if pos == index:
                return curr.data
            curr = curr.nxtaddress
            pos = pos + 1
        return 'IndexError'
    # Gives the node using the index; method used only when a Node is present Hence there is no return
    def get_node_at_index(self,index):
        curr = self.head
        counter = 0
        while curr!= None:
            if counter == index:
                return curr
            curr = curr.nxtaddress
            counter += 1
        
                
```

```python
class Dictionary:
    def __init__(self, capacity):
        self.capacity = capacity
        self.size = 0
        # creating an array of linkedList
        self.Buckets = self.make_array(self.capacity)

    def make_array(self,capacity):
        L =[]
        for i in range(capacity):
            L.append(LinkedList())
        return L
    
    def put(self,key,value):
        bucket_index = self.hash_function(key)  
        node_index  = self.get_index_of_node(bucket_index,key)
        if node_index == -1:
            # insert new node in the Bucket
            self.Buckets[bucket_index].add(key, value)
            self.size +=1
            # Set Load Factor
            load_factor = self.size/self.capacity
            print(load_factor)
            if load_factor >= 2:
                self.rehash();'''calling Rehash to resize the Bucket as Load factor exceeds '2'.'''
        else:
            # update
            node = self.Buckets[bucket_index].get_node_at_index(node_index)
            node.value = value
            
    def __setitem__(self,key,value):                                                        
        self.put(key,value)
    
    def __getitem__(self,key):
        return self.get(key)
    
    def __str__(self):
        for i in self.Buckets:
            i.traverse()
        return' '

    def get(self,key):
        bucket_index = self.hash_function(key)
        res =  self.Buckets[bucket_index].search(key)
        if res == -1:
            return 'Not Present'
        else:
            node = self.Buckets[bucket_index].get_node_at_index(res)
            return node.value

    def __delitem__(self,key):
        bucket_index = self.hash_function(key)
        self.Buckets[bucket_index].remove(key)
        self.size = self.size - 1

    def rehash(self):
        self.capacity = self.capacity *2
        new_capacity = self.capacity
        old_bucket = self.Buckets
        self.size = 0
        self.Buckets = self.make_array(new_capacity);'''resize the bucket with new capacity'''
        for i in old_bucket:
            for j in range(i.size()):
                node = i.get_node_at_index(j)
                node.key_item = node.key 
                node.value_item = node.value
                # put each old_bucket value into the resized_bucket
                self.put(node.key_item,node.value_item)

    def get_index_of_node(self,bucket_index,key):
       node_index = self.Buckets[bucket_index].search(key)
       return node_index
        
    def hash_function(self,key):
        return abs(hash(key)) % self.capacity
```
